Handle Missing Values

```python
#treating missing values

for i in Cars.columns:
    if Cars[i].dtypes=='object':
        m=Cars[i].mode()[0]
        Cars[i]=Cars[i].fillna(m)
    else:
        me=Cars[i].mean()
        Cars[i]=Cars[i].fillna(me)

```
